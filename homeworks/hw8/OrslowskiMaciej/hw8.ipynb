## 3. Wstępna analiza danych

#### Macierz korelacji

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="BrBG")
plt.show()
```

#### Histogramy

```python
df.hist(figsize=(15, 15), bins=15)
plt.show()
```

## 4. Zależności zmiennych

Po macierzy korelacji widać, że dużo zmiennych jest ze sobą powiązanych. Postanowiłem bliżej przyjrzeć się niektórym zależnościom

#### GDP per capita & Generosity

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df, x="Economy (GDP per Capita)", y="Generosity")
plt.show()
```

Jak widać, wyższe PKB na mieszkańca, które można powiązać ze średnim standardem życia mieszkańców danego kraju nie wpływa na wzrost hojności ludzi tam mieszkających. Wręcz przeciwnie, minimalne wartości hojności osiągane były w krajach z wysokim PKB na mieszkańca.

#### Freedom & Life expectancy

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df, x="Freedom", y="Health (Life Expectancy)")
plt.show()
```

Tutaj z kolei widać, że w krajach o wyższym wskaźniku wolności obywateli ci obywatele żyją śrenio dłużej.

#### Region & Trust (Government Corruption)

```python
plt.figure(figsize=(10, 20))
sns.boxplot(data=df, y="Region", x="Trust (Government Corruption)")
plt.show()
```

```python
df.loc[df["Region"] == "Middle East and Northern Africa"]\
    .sort_values("Trust (Government Corruption)", ascending=False)\
    .head(n=1)
```

Po powyższym wykresie widać, że największy poziom zaufania do swoich rządów mają obywatele w krajach Europy Zachodniej, Ameryki Północnej oraz Australii i Oceanii. W tym ostatnim wymienionym regionie zaufanie osiąga śrenio najwyższe wartości. W pozostałych częściach świata ten wskaźnik śrenio utrzymuje się na podobnym poziomie. 

Najniższą medianę wskaźnika zaufania osiągnęły kraje w centralnej i wschodniej części Europy, co stanowi duży kontrast w stosunku do zachodniej części tego kontynentu. Europa Zachodnia posiada jednak największe zróżnicowanie pod względem tej wartości. 

Co ciekawe jednak, największe zaufanie do swojego rządu mają obywatele Kataru, który leży na bliskim wschodzie.
